```python
import numpy as np
```

```python
def euclidean_distance(a, b):
    return np.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)

def generate_input_file(filename, m, n, k):
    with open(filename, 'w') as f:
        # Write grid dimensions
        f.write(f"{m} {n}\n")
        
        # Write number of obstacles
        f.write(f"{k}\n")
        
        # Generate obstacle coordinates
        obstacles = set()
        while len(obstacles) < k:
            obstacle = (np.random.randint(0, m - 1), np.random.randint(0, n - 1))
            obstacles.add(obstacle)
        
        for obstacle in obstacles:
            f.write(f"{obstacle[0]} {obstacle[1]}\n")
        
        # Generate starting cell and destination cell coordinates
        start = (np.random.randint(0, m - 1), np.random.randint(0, n - 1))
        destination = (np.random.randint(0, m - 1), np.random.randint(0, n - 1))

        while (start == destination) or (start in obstacles) or (destination in obstacles) or (euclidean_distance(start, destination) < np.sqrt(2)):
            start = (np.random.randint(0, m - 1), np.random.randint(0, n - 1))
            destination = (np.random.randint(0, m - 1), np.random.randint(0, n - 1))
        
        f.write(f"{start[0]} {start[1]}\n")
        f.write(f"{destination[0]} {destination[1]}\n")
```

```python
for i in range(1, 100):
    m = np.random.randint(5, 20)
    n = np.random.randint(5, 20)
    k = int(m  * n * 0.2)
    generate_input_file(f"test cases 2/input{i}.txt", m, n, k)
```
